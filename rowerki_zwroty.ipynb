```
# File location and type
file_location = "/FileStore/tables/cs_Zwroty.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "true"
delimiter = ","

# The applied options are for CSV files. For other file types, these will be ignored.
df = spark.read.format(file_type) \
  .option("inferSchema", infer_schema) \
  .option("header", first_row_is_header) \
  .option("sep", delimiter) \
  .load(file_location)

display(df)
```

```
dates = ["2016-03-01",
         "2016-04-01",
         "2016-05-01",
         "2016-06-01",
         "2016-07-01",
         "2016-08-01",
         "2016-09-01",
         "2016-10-01",
         "2016-11-01",
         "2017-03-01",
         "2017-04-01",
         "2017-05-01",
         "2017-06-01",
         "2017-07-01",
         "2017-08-01",
         "2017-09-01",
         "2017-10-01",
         "2017-11-01",
         "2018-03-01",
         "2018-04-01",
         "2018-05-01",
         "2018-06-01",
         "2018-07-01",
         "2018-08-01",
         "2018-09-01",
         "2018-10-01",
         "2018-11-01",
         "2019-03-01",
         "2019-04-01",
         "2019-05-01",
         "2019-06-01",
         "2019-07-01",
         "2019-08-01",
         "2019-09-01",
         "2019-10-01",
         "2019-11-01"]
columns = list(set(df.columns)-set(["_c0", "x", "y"]))
```

```
import pyspark.sql.functions as F
import pyspark.sql.types as T

d = df.select(F.col("_c0").alias("StreetName"),
              F.array(*columns).alias("arr"),
              F.col("x"),
              F.col("y"),
              F.array([F.lit(i) for i in dates]).alias("months"))
d = d.select(F.col("StreetName"),
             F.col("x"),
             F.col("y"),
             F.arrays_zip("arr", "months").alias("zipped"))
d = d.select(F.col("StreetName"),
             F.col("x"),
             F.col("y"),
             F.explode("zipped"))
d = d.select(F.col("StreetName"),
             F.col("x").cast(T.DoubleType()),
             F.col("y").cast(T.DoubleType()),
             F.col("col.arr").cast(T.IntegerType()).alias("value"),
             F.to_date(F.col("col.months")).alias("date"))
display(d)
```
