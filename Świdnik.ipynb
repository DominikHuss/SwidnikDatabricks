```
new_json = []

for street, coord in zip(streets, coords):

    x = str(coord[0])
    y = str(coord[1])
    r = requests.get('https://api.tomtom.com/traffic/services/4/flowSegmentData/absolute/13/json?point={}%2C{}&key=7BqDFDndaaS1f63jqARjB1EHmPaEIzpl'.format(x, y))
    print(r.json())
    json_data = json.loads(r.text)

    if r.status_code == 200:
        #freeFlowSpeed = json_data['flowSegmentData']['freeFlowSpeed']
        #currentSpeed = json_data['flowSegmentData']['currentSpeed']
        #print(freeFlowSpeed, currentSpeed)

        street_json = {"street": street, "coord_x": x, "coord_y": y, "data": json_data}
        new_json.append(street_json)
print(new_json)
```

```
import pyspark.sql.functions as F
import pyspark.sql.types as T
from pyspark.sql import Row

#df = spark.read.json(sc.parallelize(new_json))
df_source = spark.createDataFrame([Row(json=j, flow=j["data"]["flowSegmentData"], data=j["data"]["flowSegmentData"]["coordinates"]) for j in new_json])
df = df_source.select(F.col("json.street").alias("StreetName"),
                      F.col("json.coord_x").alias("Latitude"),
                      F.col("json.coord_y").alias("Longitude"),
                      F.col("data.coordinate").alias("Coord"),
                      F.col("flow.currentSpeed").alias("CurrentSpeed"),
                      F.col("flow.freeFlowSpeed").alias("FreeFlowSpeed"),
                      F.col("flow.currentTravelTime").alias("CurrentTravelTime"),
                      F.col("flow.freeFlowTravelTime").alias("FreeFlowTravelTime")).withColumn("MeasureDate", F.current_timestamp()).orderBy(F.col("StreetName"))
df_idx = df.rdd.zipWithIndex()
df = df_idx.toDF()
df = df.select(F.col("_1.StreetName").alias("StreetName"),
               F.col("_1.Latitude").alias("Latitude"),
               F.col("_1.Longitude").alias("Longitude"),
               F.col("_1.Coord").alias("Coord"),
               F.col("_1.CurrentSpeed").alias("CurrentSpeed"),
               F.col("_1.FreeFlowSpeed").alias("FreeFlowSpeed"),
               F.col("_1.CurrentTravelTime").alias("CurrentTravelTime"),
               F.col("_1.FreeFlowTravelTime").alias("FreeFlowTravelTime"),
               F.col("_1.MeasureDate").alias("MeasureDate"),
               F.col("_2").alias("ID"))
display(df)
```

### Measurements

```
df_measurements = df.select(F.col("ID").alias("StreetID"),
                            F.col("CurrentSpeed"),
                            F.col("FreeFlowSpeed"),
                            F.col("CurrentTravelTime"),
                            F.col("FreeFlowTravelTime"),
                            F.col("MeasureDate"))
display(df_measurements)
```
